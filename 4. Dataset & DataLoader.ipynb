```python
import numpy as np
import pandas as pd
import torch
import torchvision
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import math
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv(r'wine.csv')
df.shape
```

```python
# separate input ouput:
x = df.iloc[:, 1:].values
y = df.iloc[:, [0]].values
```

```python
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state= 42)
```

### Custome dataset class

```python
class CustomeDataset(Dataset):

    def __init__(self, featurs, labels, tr):
        self.featurs = torch.tensor(featurs, dtype = torch.float32)
        self.labels = torch.tensor(labels, dtype = torch.long)

    def __len__(self):
        return len(self.featurs)

    def __getitem__(self, index):
        return self.featurs[index], self.labels[index]
```

```python
'''object of CustomeDataset Class'''

train_dataset = CustomeDataset(x_train, y_train)
```

```python
test_dataset = CustomeDataset(x_test, y_test)
```

---
### DataLoader

```python
train_loader = DataLoader(train_dataset, batch_size= 4, shuffle= True, num_workers= 0)

test_loader = DataLoader(test_dataset, batch_size= 4)
```

### Training Loop

```python
epochs = 1

for epoch in range(epochs):

    for batch_feature, batch_labels in train_loader:
        print(batch_feature, '\n',batch_labels)
```
